```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
iris = datasets.load_iris()
X = iris.data
X = pd.DataFrame(X)
```

```python
Z = linkage(X, method="ward")
```

Les stratégies d'agrégation sont:

-single linkage: minimum distance between closest elements in clusters.

-complete linkage: maximum distance between elements in clusters.

-average linkage

-centroid method

```python
ig, ax = plt.subplots(1, 1, figsize=(12, 8))
D = dendrogram(Z, p=10, truncate_mode="lastp", ax=ax)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Number of points in node")
plt.ylabel("Distance.")
plt.show()
```

la meilleure partition est celle pour laquelle l'indice de la hiérarchie h est tel que 13<h<30 (une variation brusque)
